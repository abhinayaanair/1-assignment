# Explain the following termsg
* The difference in np.array, np.asarray and np.asanyarray 
* The difference between Deep copy and shallow copy 


```python
# Differences between np.array, np.asarray, and np.asanyarray

# np.array: 
# - Creates a new ndarray object. 
# - Always makes a copy of the input data unless the `copy=False` argument is specified.
# - Can be used to create arrays from lists, tuples, or other sequences, and specify data types and other attributes.

# np.asarray:
# - Converts the input to an ndarray.
# - Does not copy the data if the input is already an ndarray (unless a copy is requested with the `copy=True` argument).
# - Useful for ensuring that an input is converted to an ndarray without copying if it's already in that format.

# np.asanyarray:
# - Similar to np.asarray but retains subclasses of ndarray (e.g., pandas DataFrame).
# - Converts the input to an ndarray while preserving its type if it's already a subclass of ndarray.
# - Useful for maintaining the original data structure if it's a subclass.
```

```python

# Differences between Deep copy and shallow copy

# Shallow copy:
# - Creates a new object but does not create copies of nested objects; instead, it references the same nested objects as the original.
# - Changes to nested objects in the shallow copy will affect the original object, and vice versa.
# - In Python, shallow copies can be made using the `copy` module's `copy()` function or by using slicing with lists.

# Deep copy:
# - Creates a new object and recursively creates copies of all nested objects.
# - Changes to nested objects in the deep copy do not affect the original object, and vice versa.
# - In Python, deep copies can be made using the `copy` module's `deepcopy()` function.
```

```python

# Create a NumPy array with random integers between 1 and 10 of shape (5, 6)
array = np.random.randint(1, 11, size=(5, 6))

# a) Extract all even integers from the array
even_integers = array[array % 2 == 0]

# b) Extract all odd integers from the array
odd_integers = array[array % 2 != 0]

print("Original array:")
print(array)
print("\nEven integers:")
print(even_integers)
print("\nOdd integers:")
print(odd_integers)
```

```python
# Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10
array = np.random.randint(1, 11, size=(3, 3, 3))

# a) Find the indices of the maximum values along each depth level (third axis)
max_indices = np.argmax(array, axis=2)

# b) Perform element-wise multiplication between the array and itself
array_multiplied = np.multiply(array, array)

print("Original array:")
print(array)
print("\nIndices of maximum values along each depth level:")
print(max_indices)
print("\nElement-wise multiplication of the array with itself:")
print(array_multiplied)






```

```python
# Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# perform the following tasks using Matplotlib:
# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').
# b) Label the x-axis as 'Date'.
# c) Set the title of the plot as 'Temperature and Humidity Over Time'.

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Create a time-series dataset
np.random.seed(0)  # For reproducibility
dates = pd.date_range(start='2024-01-01', periods=100)
temperature = np.random.normal(loc=20, scale=5, size=100)  # Random temperatures
humidity = np.random.normal(loc=60, scale=10, size=100)    # Random humidity

df = pd.DataFrame({'Date': dates, 'Temperature': temperature, 'Humidity': humidity})

# Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes
fig, ax1 = plt.subplots()

# Plot 'Temperature' on the left y-axis
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='tab:blue')
ax1.plot(df['Date'], df['Temperature'], color='tab:blue', label='Temperature')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis for 'Humidity'
ax2 = ax1.twinx()
ax2.set_ylabel('Humidity', color='tab:orange')
ax2.plot(df['Date'], df['Humidity'], color='tab:orange', label='Humidity')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Show the plot
plt.show()
```

```python
# Create a NumPy array `data` containing 1000 samples from a normal distribution and perform the following tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.
# b) Overlay a line plot representing the normal distribution's probability density function (PDF).
# c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
# d) Set the title of the plot as 'Histogram with PDF Overlay'.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black', label='Histogram')

# Overlay a line plot for the normal distribution's PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2, label='PDF')

# Label the x-axis and y-axis
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Show the legend
plt.legend()

# Display the plot
plt.show()
```

```python
# Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants),
# add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate two random arrays
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrant for each point
quadrants = []
for xi, yi in zip(x, y):
    if xi > 0 and yi > 0:
        quadrants.append('Quadrant 1')
    elif xi < 0 and yi > 0:
        quadrants.append('Quadrant 2')
    elif xi < 0 and yi < 0:
        quadrants.append('Quadrant 3')
    else:
        quadrants.append('Quadrant 4')

# Create a DataFrame for seaborn
df = pd.DataFrame({'X': x, 'Y': y, 'Quadrant': quadrants})

# Create the scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='viridis', marker='o')

# Add legend, label the axes, and set the title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the plot
plt.show()
```

```python
# With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

# Enable Bokeh to display plots inline in a Jupyter notebook
output_notebook()

# Generate data for the sine wave function
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Create a new plot with title and axis labels
p = figure(title='Sine Wave Function', x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label='Sine Wave', line_width=2)

# Add grid lines
p.grid.grid_line_color = 'gray'
p.grid.grid_line_alpha = 0.5

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import linear_cmap
import numpy as np
import pandas as pd

# Enable Bokeh to display plots inline in a Jupyter notebook
output_notebook()

# Generate random categorical data
np.random.seed(0)
categories = [f'Category {i}' for i in range(10)]
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame
df = pd.DataFrame({'Category': categories, 'Value': values})

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a new plot with title and axis labels
p = figure(x_range=df['Category'], title='Random Categorical Bar Chart', 
           x_axis_label='Category', y_axis_label='Value', height=400, width=800)

# Add a bar renderer with color based on values
color_mapper = linear_cmap(field_name='Value', palette='Viridis256', low=min(values), high=max(values))
p.vbar(x='Category', top='Value', width=0.9, source=source, color=color_mapper, line_color='white')

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@Category"), ("Value", "@Value")]
p.add_tools(hover)

# Show the plot
show(p)
```
