4. Using the given x and y data, generate a 3D surface plot to visualize the function z=sin(sqroot(x^2+y^2))

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

fig=go.Figure()
fig.add_trace(go.Scatter3d(x=df.X,y=df.Y,z=df.Z))
```

1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python


# Prepare data
x = np.linspace(0, 10, 100)  # 100 points from 0 to 10
y = np.sin(x)  # y-values as sine of x

# Create a Bokeh figure
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)

# Add line to the figure
p.line(x, y, legend_label="sin(x)", line_width=2, color="blue")

# Show the plot
output_notebook()  # If using Jupyter Notebook
show(p)
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python

x = np.random.randint(1, 10, 30)
y = np.random.randint(1000, 10000, 30)
df = pd.DataFrame({'x': x, 'y': y})

# Create a new Bokeh figure
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)

# Add scatter points to the figure
p.scatter(df.x, df.y, size=10, color="navy", alpha=0.5)

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

p = figure(title="Fruits", x_range=fruits, width=600, height=400)

# Set axis labels
p.xaxis.axis_label = 'Fruits'
p.yaxis.axis_label = 'Counts'

# Add scatter points to the figure
p.scatter(fruits, counts, size=10, color="blue", alpha=0.5)

# Show the plot
show(p)
```

```python

data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(title="Histogram of Normally Distributed Data", 
           x_axis_label='Value', 
           y_axis_label='Frequency', 
           tools="")

# Calculate the center of the bins for the x-axis
x = 0.5 * (edges[:-1] + edges[1:])

# Add a quad glyph for the histogram
p.quad(top=hist, bottom=0, left=x - 0.2, right=x + 0.2, fill_color="blue", line_color="black")

# Show the plot
show(p)
```
