# 1. What is deep learning, and how is it connected to artificial intelligence?

Deep learning is a subset of machine learning that uses neural networks with multiple layers (deep neural networks) to model and understand complex patterns in data. It forms a part of artificial intelligence (AI), enabling systems to learn from large amounts of data and make predictions or decisions without explicit programming.

# 2. What is a neural network, and what are the different types of neural networks?
A neural network is a computational model inspired by the structure and function of biological neural networks in the brain. It consists of interconnected nodes (neurons) organized into layers.

Types:

Feedforward Neural Network (FNN): Signals flow in one direction.

Convolutional Neural Network (CNN): Specialized for image data.

Recurrent Neural Network (RNN): Processes sequential data with memory.

Transformer: Efficient for NLP tasks.

Autoencoder: For data compression and noise reduction.

Generative Adversarial Network (GAN): For generating synthetic data.

# 4. What is an activation function, and why is it essential in neural networks?
An activation function introduces non-linearity to a neural network, enabling it to model complex patterns. Without it, the network would behave like a linear regression model, limiting its learning capacity.

# 6. What is a multilayer neural network?
A multilayer neural network consists of an input layer, one or more hidden layers, and an output layer. The hidden layers allow the network to learn hierarchical and complex representations of the data.

# 8. What are some common types of loss functions?
- Mean Squared Error (MSE): For regression.
- Binary Cross-Entropy: For binary classification.
- Categorical Cross-Entropy: For multi-class classification.
- Hinge Loss: For SVM-like tasks.
- Huber Loss: For handling outliers in regression.

# 9. How does a neural network learn?
A neural network learns by updating its weights during training:

1. Compute output through forward propagation.
2. Evaluate the error using a loss function.
3. Adjust weights using backpropagation and an optimizer to minimize the loss.

# 10. What is an optimizer in neural networks, and why is it necessary?
An optimizer adjusts the network’s weights and biases to minimize the loss function. It determines how the network converges during training.

# 11. Could you briefly describe some common optimizers?
- SGD (Stochastic Gradient Descent): Basic method with learning rate control.
- Momentum: Accelerates convergence by considering past gradients.
- Adam: Combines momentum and adaptive learning rates (most common).
- RMSprop: Focuses on controlling step size for consistent training.

# 12. Can you explain forward and backward propagation in a neural network?
- Forward Propagation: Passes inputs through the network to compute predictions.
- Backward Propagation: Computes gradients of the loss function w.r.t. weights using the chain rule and updates them.


# 13. What is weight initialization, and how does it impact training?
Weight initialization assigns initial values to the weights in a network. Proper initialization prevents issues like vanishing/exploding gradients and speeds up convergence.



# 14. What is the vanishing gradient problem in deep learning?
In deep networks, gradients become extremely small in earlier layers, preventing them from learning effectively. This often occurs with activation functions like Sigmoid or Tanh.

# 15. What is the exploding gradient problem?
The exploding gradient problem happens when gradients grow excessively during backpropagation, causing unstable training. It often occurs in RNNs or deep networks without gradient clipping.

```python
# 6. How do you manually implement forward propagation in a simple neural network!

import math

class Perceptron:
    def __init__(self, num_inputs, thre=0.5):
        self.weights = [0.0] * num_inputs  # Initialize weights to 0
        self.bias = 0.0                    # Initialize bias to 0
        self.thre = thre                   # Classification threshold
    
    def predict(self, inputs):
        # Compute weighted sum
        weighted_sum = sum(w * x for w, x in zip(self.weights, inputs)) + self.bias
        # Sigmoid activation
        output = 1 / (1 + math.exp(-weighted_sum))  # Correct sigmoid formula
        # Apply threshold
        return 1 if output > self.thre else 0
    
    def train(self, training_data, targets, learning_rate, epochs):
        for epoch in range(epochs):
            print(f"Epoch {epoch + 1}/{epochs}")
            for input, target in zip(training_data, targets):
                # Forward pass: make a prediction
                prediction = self.predict(input)
                
                # Calculate error
                error = target - prediction
                
                # Update weights and bias
                for i in range(len(self.weights)):
                    self.weights[i] += learning_rate * error * input[i]
                self.bias += learning_rate * error  # Bias update
                
                # Print details for debugging
                print(f"Input: {input}, Target: {target}, Prediction: {prediction}, Error: {error}")
            print(f"Weights: {self.weights}, Bias: {self.bias}\n")
    
if __name__ == "__main__":
    # Initialize perceptron
    perceptron = Perceptron(num_inputs=2, thre=0.5)
    
    # Training data (AND logic gate example)
    training_data = [
        [0, 0],
        [0, 1],
        [1, 0],
        [1, 1]
    ]
    # Targets: 0 for false, 1 for true
    targets = [0, 0, 0, 1]
    
    # Hyperparameters
    learning_rate = 0.1
    epochs = 5
    
    # Train the perceptron
    perceptron.train(training_data, targets, learning_rate, epochs)
    
    # Test the perceptron
    for inputs in training_data:
        prediction = perceptron.predict(inputs)
        print(f"Inputs: {inputs}, Prediction: {prediction}")
```

```python
#8 How can you visualize the training process with accuracy and loss curves!
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Create a simple model
model = Sequential([
    Dense(128, activation='relu', input_shape=(20,)),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Generate some random data for demonstration
import numpy as np
X_train = np.random.rand(1000, 20)
y_train = np.random.randint(0, 2, 1000)
X_val = np.random.rand(200, 20)
y_val = np.random.randint(0, 2, 200)

# Train the model
history = model.fit(
    X_train, y_train,
    validation_data=(X_val, y_val),
    epochs=20,
    batch_size=32
)

# Plot accuracy and loss curves
def plot_training_curves(history):
    # Extract metrics from history
    acc = history.history['accuracy']
    val_acc = history.history['val_accuracy']
    loss = history.history['loss']
    val_loss = history.history['val_loss']
    epochs = range(1, len(acc) + 1)
    
    # Plot Accuracy
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs, acc, label='Training Accuracy', marker='o')
    plt.plot(epochs, val_acc, label='Validation Accuracy', marker='o')
    plt.title('Accuracy Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    
    # Plot Loss
    plt.subplot(1, 2, 2)
    plt.plot(epochs, loss, label='Training Loss', marker='o')
    plt.plot(epochs, val_loss, label='Validation Loss', marker='o')
    plt.title('Loss Over Epochs')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    plt.tight_layout()
    plt.show()

# Call the plotting function
plot_training_curves(history)
```

```python
# 9. How can you use gradient clipping in Keras to control the gradient size and prevent exploding gradients

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

# Create a simple model
model = Sequential([
    Dense(64, activation='relu', input_shape=(100,)),
    Dense(32, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model with gradient clipping by value
optimizer = Adam(clipvalue=1.0)  # Clip gradients to max value of 1.0
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Train the model with some random data
import numpy as np
X_train = np.random.rand(1000, 100)
y_train = np.random.randint(0, 2, 1000)

model.fit(X_train, y_train, epochs=5, batch_size=32)
```

```python
# 10.  How can you create a custom loss function in Keras!
from tensorflow.keras.utils import plot_model
from tensorflow.keras.optimizers import Adam

def custom_loss(y_act,y_pred):
    mae=tf.redice_mean(tf.abs(y_act-y_pred))
    mse=tf.redice_mean(tf.square(y_act-y_pred))
    return mae+mse

model=Sequential([
    Dense(14,activation='relu',input_shape=(100,)),
    Dense(7,activation='relu'),
    Dense(1)
])

model.compile(optimizer=Adam(),loss=custom_loss)

plot_model(model,show_shapes=True,show_dtype=True,to_file='model_structure.png')
```
